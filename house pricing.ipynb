```python
features = df[['Rooms','Distance','Bedroom2','Bathroom','Car','Landsize','BuildingArea','YearBuilt','Lattitude','Longtitude']]
label = df['Price']
```

impute values in the null rows

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer()
```

```python
cols_to_impute = [col for col in features if features[col].isnull().any()]
```

```python
cols_to_impute
```

```python
features_Imp = pd.DataFrame(imputer.fit_transform(features))
```

```python
features_Imp.columns = features.columns
```

```python
from sklearn.model_selection import train_test_split
```
